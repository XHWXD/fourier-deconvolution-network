```python
from skimage.io import imread
from fdn_predict import show
```

# Example 1

- Finedtuned model from paper applied to color image

```python
run fdn_predict.py \
--image data/example1.png \
--kernel data/example1.dlm \
--sigma 2.55 \
--n-stages 10 \
--finetuned 1 \
--output _deblurred.png \
--model models/sigma_1.0-3.0
```

```python
show(imread('data/example1.png'),figsize=(12,12),title='blurred input')
show(imread('_deblurred.png'),figsize=(12,12),title='deblurred output')
```

# Example 2

- Greedy model for larger noise level range (cf. supplemental material) applied to grayscale image

```python
run fdn_predict.py \
--image data/example2.png \
--kernel data/example2.dlm \
--sigma 1.5 \
--n-stages 10 \
--finetuned 0 \
--output _deblurred.png \
--model models/sigma_0.1-12.75
```

```python
show(imread('data/example2.png'),figsize=(12,12),title='blurred input')
show(imread('_deblurred.png'),figsize=(12,12),title='deblurred output')
```
